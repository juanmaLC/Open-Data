# Borrar datasets

```python
#Borrar todos los datasets de localhost 
ckan =ckanapi.RemoteCKAN('http://opendata-test.ugr.es:5000', apikey=APIKEY)

bp = {
    'include_private': True,
    'rows':100
    
}


listapaquetes=ckan.action.package_search(**bp)


for paquete in listapaquetes['results']:
    dict_p ={
        'id': paquete['id']

    }

    ckan.action.package_delete(**dict_p)
    ckan.action.dataset_purge(**dict_p)
```

# Create investiga.ugr.es Personal investigador

```python
APIKEY='1111'
```

```python

ckan= ckanapi.RemoteCKAN("https://opendata.ugr.es",apikey=APIKEY)


url='https://investiga.ugr.es/wp-json/datos-abiertos/v1/investigadores/'

for i in range(46):
    contador=i+1
    urlFinal=url+str(contador)
    nameFinal='Personal Investigador ' + str(contador)


    resource=ckan.action.resource_create(
        package_id='6969a425-dc9d-474f-8c06-9307a8435d78',
        url=urlFinal,
        format='JSON',
        name=nameFinal
    )
    ckan.action.resource_view_create(
        resource_id=resource['id'],
        title='Personal investigador de la Universidad de Granada',
        view_type='webpage_view'
    )
```
